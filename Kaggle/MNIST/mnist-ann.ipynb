```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import seaborn as sns
```

```python
df = pd.read_csv('../input/digit-recognizer/train.csv')
y_train = df['label'].values
X_train = df.drop(['label'],axis=1).values
```

```python
X_test = pd.read_csv('../input/digit-recognizer/test.csv').values
X_test[0]
```

```python
X_train[0]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    tf.keras.layers.Dense(10,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model.fit(X_train,y_train,epochs=20)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_labels = [np.argmax(i) for i in y_pred]
```

```python
y_pred_labels = np.array(y_pred_labels)
y_pred_labels[:10]
```

```python
ImageId = np.array(range(1,28001))
ImageId[-1]
```

```python
col = np.concatenate((ImageId[:,np.newaxis],y_pred_labels[:,np.newaxis]),axis=1)
```

```python
col[:10]
```

```python
submission_df = pd.DataFrame(col,columns=['ImageId','Label'],index=None)
```

```python
submission_df.head()
```

```python
pd.DataFrame(submission_df).to_csv('./submission.csv',columns=['ImageId','Label'],index=None)
```
